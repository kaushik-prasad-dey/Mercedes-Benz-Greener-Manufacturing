**K-fold Cross Validation using xgboost**

k fold cross validation is a technique used in ML to assess the performance of a model and reduce the risk of overfitting.

The dataset is split into k folds and the model is trained and evaluted k times and each time using a different fold as the test set and the remaining fold as the training set.

The performance metrics are the averaged over the k ierations.

```python
#you have to create the k fold cross validation with xgb
dmatrix_train = xgb.DMatrix(data=train_data_feature_trans, label=train_data_target)
params ={
    'objective':'reg:linear',
    'colsample_bytree':0.3,
    'learning_rate':0.3
}
model_cv=xgb.cv(dtrain=dmatrix_train, params=params, nfold=3, num_boost_round=50,early_stopping_rounds=10)
```

```python
print(model_cv)
```

```python
#i want to see the cross validation table last 2 records
print(model_cv.tail(2))
```

By using k-fold cross validation , RMSE is reduced approxmately 10%, RMSE=11
